```python
yaml_content = """path: /content/final_dataset/final_dataset
train: train
val: train

names:
  0: reusable
  1: disposable
  2: others
"""

with open("/content/final_dataset/final_dataset/data.yaml", "w") as f:
    f.write(yaml_content)

print("✅ data.yaml 경로 수정 완료")
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")
model.train(
    data="/content/final_dataset/final_dataset/data.yaml",
    epochs=50,
    imgsz=640,
    name="train_custom3"
)

# 학습된 모델 저장
model_path = "/content/custom_yolo.pt"
model.save(model_path)
print(f"✅ 모델 저장 완료: {model_path}")
```
